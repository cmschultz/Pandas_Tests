### Importing Pandas Library

```python
import pandas as pd
```

### Types in Pandas

One type in pandas is Series. 
It is one dimensional and you pass a list in it.

```python
# They have only one dimension
series = pd.Series(["BMW", "Toyota", "Honda"])
series
```

```python
colours = pd.Series(["Red", "Blue", "White"])
colours
```

The other type of data is the Dataframe. It has two dimensions and recives a Dictionary for the construction.

```python
# Dataframe __ They have two dimensions

car_data = pd.DataFrame({"Car make": series, "Colour": colours})
car_data
```

When you see axis = 0 you are talking about a row. When you say axis = 1 you are talking about a column.

### Importing .csv Files

You can import an csv file using the function ps.read_csv(), passing the archieve name with the .csv and between ".

```python
# Import data
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

### Exporting data to .csv files

Use .to_csv() function to get a csv version of the dataset. Notice that you have to put an index = false, this is because every time pandas read a csv file, it make a new index. If you export the csv with the index, you will have two identical indexes when you read the csv back to a dataframe.

```python
# Run this cell only one time. 
car_sales.to_csv("exported-car-sales.csv", index = False) #this index false make you dont have to columns index when you create a new dataframe importing it.
```

```python
exported_car_sales = pd.read_csv("exported-car-sales.csv")
exported_car_sales
```
